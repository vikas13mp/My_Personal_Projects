```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/rajtilakls2510/car_price_predictor/master/Cleaned_Car_data.csv')
```

```python
df
```

```python
df.drop(columns = ['Unnamed: 0'], axis =1 ,inplace = True)
```

```python
df = df[df['Price']<6e6].reset_index(drop=True)
```

```python
df
```

```python
x = df.drop(columns=['Price'])
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python

ohe.categories_
```

```python
column_train = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['name','company','fuel_type']), remainder= 'passthrough')
```

```python
print(column_train)
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_train,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
import numpy as np
```

```python
scores = []
for i in range(1000):
  X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = i)
  lr = LinearRegression()
  pipe = make_pipeline(column_train,lr)
  pipe.fit(X_train,y_train)
  y_pred = pipe.predict(X_test)
  scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 661)
lr = LinearRegression()
pipe = make_pipeline(column_train,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
